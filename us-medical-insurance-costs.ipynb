## Brief Overview

* Step 1: Import the relevant libraries and read the CSV file.
* Step 2: Find the average age, bmi, number of children and insurance charge.
* Step 3: Find the range for age, bmi, number of children and insurance charge.
* Step 4: Plot the distribution of males and females in the dataset.
* Step 5: Plot the distribution of people in their respective regions.

```python
# Step 1: Import the relevant libraries and read the CSV file.

import csv
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv('insurance.csv')
print(df)
```

```python
# Step 2: Find the average age, bmi, number of children and insurance charge.

avg_age = df['age'].mean()
avg_bmi = df['bmi'].mean()
avg_children = df['children'].mean()
avg_charge = df['charges'].mean()

print(f"The average age is {avg_age}.")
print(f"The average bmi is {avg_bmi}.")
print(f"The average number of children is {avg_children}.")
print(f"The average insurance charge is {avg_charge}.")
```

```python
# Step 3: Find the range for age, bmi, number of children and insurance charge.

max_age = df['age'].max()
min_age = df['age'].min()
print(f"The maximum age is {max_age} and the minimum age is {min_age}.")

max_bmi = df['bmi'].max()
min_bmi = df['bmi'].min()
print(f"The maximum BMI is {max_bmi} and the minimum BMI is {min_bmi}.")

max_children = df['children'].max()
min_children = df['children'].min()
print(f"The maximum number of children is {max_children} and the minimum number of children is {min_children}.")

max_charge = df['charges'].max()
min_charge = df['charges'].min()
print(f"The maximum insurance charge is {max_charge} and the minimum insurance charge is {min_charge}.")
```

```python
# Step 4: Plot the distribution of males and females in the dataset.

gender_split = df['sex'].value_counts()
print(gender_split)

df['sex'].value_counts().plot(kind='bar')
```

```python
# Step 5: Plot the distribution of people in their respective regions.

location_split = df['region'].value_counts()
print(location_split)

df['region'].value_counts().plot(kind='bar')
```

## Disover Correlation

* Step 1: Find the correlation between insurance charge and age.

```python
# Step 1: Find the correlation between insurance charge and age.

df.plot(x='age', y='charges', kind='scatter')
```

```python
# Step 2: Find the correlation between insurance charge and sex.

df.groupby('sex').charges.mean().plot(kind='bar')
```

```python
# Step 3: Find the correlation between insurance charge and bmi.

df.groupby('bmi').charges.mean().plot(kind='area')
```

```python
# Step 4: Find the correlation between insurance charge and number of children.

df.groupby('children').charges.mean().plot(kind='bar')
```

```python
# Step 5: Find the correlation between insurance charge and smoker status.

df.groupby('smoker').charges.mean().plot(kind='bar')
```

```python
# Step 6: Find the correlation between insurance charge and region.

df.groupby('region').charges.mean().plot(kind='bar')
```

## Understand Correlation

* Positive correlation: There is a relationship between two variables that tend to move in the same direction.
* The following factors are correlated with insurance charges: sex, BMI, smoker status, and region.
* As BMI increases, insurance charges also rises. 
* Insurance charges are also higher among males or smokers, compared to females and non-smokers.
* Also, insurance charge is higher in Southeast region, followed by Northeast and then Northwest and Southwest.
* There is no correlation between insurance charge and the number of children.

## Analyze & Justify

* Among the five factors listed above, the correlation between insurance charges and sex and region seem to be the most intriguing.
* The questions we are trying to answer are: Why are insurance charges higher among males? Also, why do people in the Southeast region have the highest average insurance charge compared to the other regions?
* We are narrowing our focus down to these two factors because a) there is no correlation between insurance charges and the number of children and b) the reason insurance charges is correlated with BMI and smoker status is because a high BMI or smoking habit leads to increased risk of developing health conditions, thus increasing claims on healthcare.
* It is also worthwhile to note that there is a positive correlation between BMI and obesity. According to CDC.gov, when BMI is greater than or equal to 30 kg/m^2, the person is considered obese.
* Step 1: Insurance charges might be higher among males because there are more males who smoke, compared to females.
* Step 2: Insurance charges might be higher among males because there are more obese males, compared to females.
* Step 3: Insurance charges might be higher in the Southeast region because there are more smokers in that region.
* Step 4: Insurance charges might be higher in the Southeast region because there are more obese people in that region.

```python
# Step 1: Insurance charges might be higher among males because there are more males who smoke, compared to females.

num_male_smokers = len(df.loc[(df['sex'] == 'male') & (df['smoker'] == 'yes')])
num_female_smokers = len(df.loc[(df['sex'] == 'female') & (df['smoker'] == 'yes')])

print(f"The number of male smokers is {num_male_smokers}, while the number of female smokers is {num_female_smokers}.")
```

```python
# Step 2: Insurance charges might be higher among males because there are more obese males, compared to females.

num_obese_males = len(df.loc[(df['sex'] == 'male') & (df['bmi'] > 30)])
num_obese_females = len(df.loc[(df['sex'] == 'female') & (df['bmi'] > 30)])
print(f"The number of obese males is {num_obese_males}, while the number of obese females is {num_obese_females}.")
```

```python
# Step 3: Insurance charges might be higher in the Southeast region because there are more smokers in that region.

num_smokers_southeast = len(df.loc[(df['smoker'] == 'yes') & (df['region'] == 'southeast')])
num_smokers_northeast = len(df.loc[(df['smoker'] == 'yes') & (df['region'] == 'northeast')])
num_smokers_northwest = len(df.loc[(df['smoker'] == 'yes') & (df['region'] == 'northwest')])
num_smokers_southwest = len(df.loc[(df['smoker'] == 'yes') & (df['region'] == 'southwest')])

print(f"The number of smokers according to regions are {num_smokers_southeast} in Southeast, {num_smokers_northeast} in Northeast,") 
print(f"followed by {num_smokers_northwest} in Northwest and {num_smokers_southwest} in Southwest.")
```

```python
# Step 4: Insurance charges might be higher in the Southeast region because there are more obese people in that region.

num_obese_southeast = len(df.loc[(df['bmi'] > 30) & (df['region'] == 'southeast')])
num_obese_northeast = len(df.loc[(df['bmi'] > 30) & (df['region'] == 'northeast')])
num_obese_northwest = len(df.loc[(df['bmi'] > 30) & (df['region'] == 'northwest')])
num_obese_southwest = len(df.loc[(df['bmi'] > 30) & (df['region'] == 'southwest')])

print(f"The number of obese people according to regions are {num_obese_southeast} in Southeast, {num_obese_northeast} in Northeast,") 
print(f"Followed by {num_obese_northwest} in Northwest and {num_obese_southwest} in Southwest.")
```

## Conclusion

* There is no correlation between the number of children and insurance charges.
* Insurance charge is correlated with BMI and smoking status due to increasing health risks.
* Also, males have higher insurance charges compared to females because there are more males who smoke or have a higher BMI and are considered obese.
* In terms of region, the Southeast region has the highest average insurance charge, followed by the Northeast region, Northwest region and Southwest region. This can be attributed to the number of people who smoke or have a high BMI.
* However, there is some discrepency between using BMI levels to explain the relationship between region and insurance charges. For instance, the Southwest region has the lowest average insurance charge, but yet they have the second highest number of obese people (171). Further analysis is required here.
